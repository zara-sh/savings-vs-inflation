```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Quarter': [
        'Mar-16','Jun-16','Sep-16','Dec-16',
        'Mar-17','Jun-17','Sep-17','Dec-17',
        'Mar-18','Jun-18','Sep-18','Dec-18',
        'Mar-19','Jun-19','Sep-19','Dec-19',
        'Mar-20','Jun-20','Sep-20','Dec-20',
        'Mar-21','Jun-21','Sep-21','Dec-21',
        'Mar-22','Jun-22','Sep-22','Dec-22',
        'Mar-23','Jun-23','Sep-23','Dec-23',
        'Mar-24','Jun-24','Sep-24','Dec-24',
        'Mar-25','Jun-25'
    ],
    'Quarterly_CPI_Change': [
        -0.2,0.4,0.7,0.5,
        0.5,0.2,0.6,0.6,
        0.4,0.4,0.4,0.5,
        0.0,0.6,0.5,0.7,
        0.3,-1.9,1.6,0.9,
        0.6,0.8,0.8,1.3,
        2.1,1.8,1.8,1.9,
        1.4,0.8,1.2,0.6,
        1.0,1.0,0.2,0.2,
        0,0
    ]
}

df = pd.DataFrame(data)
# Nominal savings per quarter ($5/day Ã— 91 days)
df['Savings_per_Quarter'] = 5* 91

# Check your DataFrame columns
print(df.columns)
```

```python
df
```

```python
df["Quarterly_CPI_Change_khar"]= df["Quarterly_CPI_Change"] *1554
```

```python
df
```

```python
# Real savings = Nominal / (1 + CPI/100)
df['Real_Savings'] = df['Savings_per_Quarter'] / (1 + df['Quarterly_CPI_Change']/100)

# Cumulative sums
df['Cumulative_Nominal'] = df['Savings_per_Quarter'].cumsum()
df['Cumulative_Real'] = df['Real_Savings'].cumsum()

df
```

```python
plt.figure(figsize=(14,7))

# Plot nominal and real cumulative savings
plt.plot(df['Quarter'], df['Cumulative_Nominal'], marker='o', color='blue', label='Cumulative Nominal')
plt.plot(df['Quarter'], df['Cumulative_Real'], marker='o', color='green', label='Cumulative Real (Adjusted for Inflation)')

# Highlight difference with shaded area
plt.fill_between(df['Quarter'], df['Cumulative_Nominal'], df['Cumulative_Real'], 
                 color='orange', alpha=0.2, label='Inflation Impact')

# Make x-axis readable
plt.xticks(rotation=45)
plt.title('$5/day Savings: Nominal vs Real Value (2016â€“2025)', fontsize=16)
plt.xlabel('Quarter', fontsize=12)
plt.ylabel('Savings ($)', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
